## Query 2: Top 10 Products By Store

```python
collection = db.top_3_products_by_store
df = pd.DataFrame(list(collection.find()))

exploded_df = df.explode('topProducts')

exploded_df['productName'] = exploded_df['topProducts'].apply(lambda x: x['name'])
exploded_df['totalSales'] = exploded_df['topProducts'].apply(lambda x: x['totalSales'])

pivot_df = exploded_df.pivot(index='_id', columns='productName', values='totalSales')
pivot_df = pivot_df[['laptop', 'backpack', 'pens']]
pivot_df.reset_index(inplace=True)

pivot_df = pivot_df.rename(columns={'_id': 'Location', 'laptop': '1st - Laptop', 'backpack': '2nd - Backpack', 'pens': '3rd - Pens'})
print(pivot_df.to_string(index=False))
```

## Query 3: Store Rankings

```python
collection = db.store_rankings
df = pd.DataFrame(list(collection.find()))

df = df.rename(columns={'_id': 'Location'})
print(df.to_string(index=False))
```

## Purchased Methods By Gender

```python
collection = db.purchased_methods_by_gender
df = pd.DataFrame(list(collection.find()))
exploded_df = df.explode('purchasedMethods')

exploded_df['method'] = exploded_df['purchasedMethods'].apply(lambda x: x['method'])
exploded_df['count'] = exploded_df['purchasedMethods'].apply(lambda x: x['count'])

pivot_df = exploded_df.pivot(index='_id', columns='method', values='count')
pivot_df.reset_index(inplace=True)
pivot_df = pivot_df.rename(columns={'_id': 'Gender', 'In store': 'In Store', 'Online': 'Online', 'Phone': 'Phone'})

print(pivot_df.to_string(index=False))
```

## Monthly Sales

```python
collection = db.yearly_monthly_total_sales
df = pd.DataFrame(list(collection.find()))

exploded_df = df.explode('monthlySales')
exploded_df['month'] = exploded_df['monthlySales'].apply(lambda x: x['month'])
exploded_df['totalSales'] = exploded_df['monthlySales'].apply(lambda x: x['totalSales'])
exploded_df['month'] = exploded_df['month'].astype(str)

pivot_df = exploded_df.pivot(index='_id', columns='month', values='totalSales')
pivot_df.reset_index(inplace=True)
pivot_df = pivot_df.rename(columns={'_id': 'Year', '1': 'Jan', '2': 'Feb', '3': 'Mar', '4': 'Apr', '5': 'May', '6': 'Jun', '7': 'Jul', '8': 'Aug', '9': 'Sep', '10': 'Oct', '11': 'Nov', '12': 'Dec'})

print(pivot_df.to_string(index=False))
```
